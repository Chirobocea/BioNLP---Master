```python
import csv

# Open the CSV file
with open('/kaggle/input/sentimental-analysis-for-tweets/sentiment_tweets3.csv', newline='') as csvfile:
    # Create a CSV reader
    reader = csv.DictReader(csvfile)
    
    # Initialize an empty list to store the comments
    comments = []
    
    # Iterate over each row in the CSV
    for row in reader:
        # Append the value of the "Comment" column to the comments list
        comments.append(row['message to examine'])

# Print the list of comments
print(comments[:5])
```

```python
from gritlm import GritLM
import torch
import numpy as np
from tqdm import tqdm

torch.backends.cuda.enable_mem_efficient_sdp(False)
torch.backends.cuda.enable_flash_sdp(False)

# Loads the model for both capabilities; If you only need embedding pass mode="embedding" to save memory (no lm head)
model = GritLM("GritLM/GritLM-7B", torch_dtype="auto")
batch_size = 8
all_enc = []

### Embedding/Representation ###
def gritlm_instruction(instruction):
    return "<|user|>\n" + instruction + "\n<|embed|>\n" if instruction else "<|embed|>\n"

# No need to add instruction for retrieval documents
for start in tqdm(range(0, len(comments), batch_size)):
    if start+batch_size > len(comments):
        stop = len(comments)
    else:
        stop = start+batch_size
    enc = model.encode(comments[start:stop], instruction=gritlm_instruction(""))
    all_enc.extend(enc)

        
# Save the NumPy array as a .npy file
np.save('depression_embed.npy', np.array(all_enc))
```
